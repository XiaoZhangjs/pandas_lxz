```python
''' 1、数据集基本信息 '''
import pandas as pd
df = pd.read_csv('../data/taitanni.csv')
# 探索行(样本数)/列(特征数)
num1, num2, num3 = df.shape, df.shape[0], df.shape[1]
# 探索索引和列名
num4, num5 = df.index, df.columns
# 将索引和列名转换成列表
num6, num7 = num4.tolist(), list(num5)
# 探索数据类型
num8 = df.dtypes
num1, num2, num3, num4, num5, num6[:5], num7[:5], num8
```

```python
''' 2、汇总统计 '''
# 求和 
num1, num2 = df.sum(), df['Age'].sum()
# 累加
num3 = df['Age'].cumsum()
# 最大值/最小值
num4, num5 = df.max(), df['Age'].min()
# 平均值/中位数
num6, num7 = df['Age'].mean(), df.median()
# 基本数据统计
num8, num9 = df.describe().round(3), df['Age'].describe().T
# 保留小数点
num10 = round(df.describe(), 2)
'【num1】',num1, '【num2】',num2, '【num3】',num3.head(), '【num4】',num4, '【num5】',num5,\
'【num6】',num6, '【num7】',num7, '【num8】',num8, '【num9】',num9, '【num10】',num10,
```
