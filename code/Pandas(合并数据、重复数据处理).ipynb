```python
import pandas as pd
data_1 = {
        'id': ['1', '2', '3', '4', '5'],
        '名字': ['小万', '小猫', '小龙', '小梁', '小妹'], 
        '小名': ['旺旺', '喵喵', '二狗', '眼镜', '米子']}

data_2 = {
        'id': ['4', '5', '6', '7', '8'],
        '名字': ['小李', '小丽', '小华', '小韩', '小文'], 
        '小名': ['效力', '粒子', '花花', '劳资', '小雯']}

data_3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10'],
        'score': [43, 35, 15, 26, 10, 47, 67, 1, 78]}

df1 = pd.DataFrame(data_1, columns=['id', '名字', '小名'])
df2 = pd.DataFrame(data_2, columns=['id', '名字', '小名'])
df3 = pd.DataFrame(data_3, columns=['id', 'score'])
?df1.iloc()
```

```python
''' 1、以行和列的维度来合并 '''
df_0 = pd.concat([df1, df2])  # 默认以行合并
df_1 = pd.concat([df1, df2], axis=1)  # 以列的维度合并
print('【df_0】\n', df_0, '\n【df_1】\n', df_1)
```

```python
''' 
2、按照指定的值合并 

参数介绍
-------------
一般要设置下面这 4 个参数
left: 拼接的左侧DataFrame对象
right: 拼接的右侧DataFrame对象
on: 要指定的连接键
how：（没有指定 how 方式时，默认inner）
inner 是取 on 对应连接键的交集，outer 取 on 对应的并集；
left 只使用 on 对应连接键左帧的键,right 只使用右帧的键
'''
pd.merge(df1, df2, on='id')  # 没有设置 how ，采用默认的 inner,取交集
```
